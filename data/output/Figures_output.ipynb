```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from scipy.signal import savgol_filter
```

### Load cloud cover

```python
Goonhilly = pd.read_csv("cloud_cover/Goonhilly_eumetsat_2023-06-01_2024-06-02_detailed_df.csv",delimiter=",", dtype=str)
plt.figure(figsize=(14,8))
plt.plot(pd.to_datetime(Goonhilly['time'][:]), Goonhilly['cloud_cover'].astype(float))
plt.plot(pd.to_datetime(Goonhilly['time'][:]), savgol_filter(Goonhilly['cloud_cover'].astype(float), 500, 3))
plt.grid()
plt.xlabel('Time')
plt.ylabel('Cloud cover')
plt.title('Goonhilly cloud cover')
```

```python
df = Goonhilly[['cloud_cover']]
df = df.rename(columns={'cloud_cover': 'Goonhilly'})
df.insert(1, 'Madrid', Madrid['cloud_cover'])
df.insert(2, 'Nemea', Nemea['cloud_cover'])
df.insert(3, 'Oberpfaffenhofen', Oberpfaffenhofen['cloud_cover'])
df.insert(4, 'Tenerife', Tenerife['cloud_cover'])
df.insert(5, 'TheHague', TheHague['cloud_cover'])
df.insert(6, 'Trauen', Trauen['cloud_cover'])
```

```python
df
```

```python
corr_map = df.corr(method='pearson')
```

```python
import seaborn as sns
mask = np.zeros_like(corr_map, dtype=bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr_map,  mask=mask, annot = True)

plt.savefig('pearson_corr.svg')
plt.show()
```
